### Problem 1.1 (1 pts)
- (1 pts) For $1 \leq p \leq q \leq \infty$, show that there always exists a constant $C = C(n,p,q) > 0$ such that $\|x\|_p \leq C\|x\|_q$ for every $x \in \mathbb{R}^n$

**Solution:**

There is exists so called ["Hölder's inequality"](https://en.wikipedia.org/wiki/Hölder%27s_inequality). It says that for $ (S, \Sigma, \mu) $ a measure space and let $ p, q \in [1, \infty] $ with $ 1/p + 1/q = 1 $. Then for all measurable real- or complex-valued functions $ f $ and $ g $ on $ S $,
$$ \left\| fg \right\| \le \left\| f \right\|_p \left\| g \right\|_q$$

Using this inequality and the definition of p-norm, we can write:
$$ \left\| x \right\|_p = \left( \sum_{i=1}^{n} \left| x_i \right|^p * 1\right)^{1/p} \le \left( \left( \sum_{i=1}^{n} \left| x_i \right|^{pr}\right)^{1/r} \left( \sum_{i=1}^{n} 1^s \right)^{1/s} \right)^{1/p} = \left( \sum_{i=1}^{n} \left| x_i \right|^q\right)^{1/q} \cdot n^{(q - p)/(pq)} = \left\| x \right\|_q \ \cdot n^{1/p - 1/q}$$ 

Where, $ 1/r = p/q$ and $ 1/s = (q - p)/q $. And we can denote $ C = n^{(q - p)/(pq)} $. Then we can rewrite: 
$$ \left\| x \right\|_p  \le C \left\| x \right\|_q$$


### Problem 1.2 (4 pts)

For any norm $\|\cdot\|$ in finite dimentional space $\mathbb{C}^{N}$ we define dual norm $\|x\|^{'} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\}$, where ${\cdot}^{\star}$ denotes transpose complex-conjugate vector.
+ (2 pts). Prove that $\|\cdot\|^{'}$ is a norm.
+ (2 pts). Find dual norm for $l_p$ norm

**Solution:**

By the definition from Lecture 1, we can call a function norm, if it's satisfying all of the following:
- Positivity: $\Vert x \Vert = 0$ then $x = 0$ ;
- Homogeneity: $\Vert \alpha x \Vert = |\alpha| \Vert x \Vert$ ;
- Triangle inequality: $\Vert x + y \Vert \leq \Vert x \Vert + \Vert y \Vert$ . 

Let's go through all points step-by-step:

1. Positivity:

By the definition we have: 
$$\|x\|^{'} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\}$$

Since $ \left| y^{\star} x \right| $ is the absolute value of an inner product, it is always non-negative:
$$ \left\| x \right\|^{'} \ge 0$$

Now, we suppose that $\left\| x \right\|^{'} = 0$. It means:
$$\sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\} = 0$$

This possible if and only if $\left| y^{\star} x \right| = 0$ for all $y$ with $\left\|y\right\|=1$. And it leads us to conclusion, that $y^{\star} x = 0$ for all $y$ with $\left\|y\right\|=1$. Since $y$ can be any vector on the unit sphere in $ \mathbb{C}^N $, this can only happen if $ x=0 $.

2. Homogeneity:

Let $\alpha \in \mathbb{C}$ be any scalar, and let $x \in \mathbb{C}^N$. By the definition of dual norm we have: 
$$\|x\|^{'} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\}$$

Using [Multiplicativity](https://en.wikipedia.org/wiki/Absolute_value) property of absolute value and inner product, we can write:
$$\|\alpha x\|^{'} = \sup\left\{\left|y^{\star} (\alpha x)\right|:\left\|y\right\|=1\right\}$$
$$\|\alpha x\|^{'} = \sup\left\{\left| \alpha \right| \cdot \left|y^{\star} x\right|:\left\|y\right\|=1\right\}$$

Whereas $\left| \alpha \right|$ is a constant, it could be written as:
$$\|\alpha x\|^{'} = \left| \alpha \right| \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\} = \
                    \left| \alpha \right| \left\| x \right\|^{'}$$

3. Triangle Inequality:

Assume $x, z \in \mathbb{C}^N$. We need to show:
$$\left\| x + z \right\|^{'} \le \left\| x \right\|^{'} + \left\| z \right\|^{'}$$

let's use the definition of a dual norm and write down an expression for $x + z$:
$$\|x + z\|^{'} = \sup\left\{\left|y^{\star} (x + z)\right|:\left\|y\right\|=1\right\}$$
Using linearity property of inner product, we can rewrite this as:
$$\|x + z\|^{'} = \sup\left\{\left|y^{\star}x + y^{\star}z\right|:\left\|y\right\|=1\right\}$$ 
Recall that our $x, z \in \mathbb{C}^N$. Applying the triangle inequality for complex numbers ($\left|a + b \right| \le \left| a \right| + \left| b \right|$), we can write:
$$\|x + z\|^{'} = \sup\left\{\left|y^{\star}x \right| + \left| y^{\star}z\right|:\left\|y\right\|=1\right\}$$
And using [property of supremum](https://www.math.ucdavis.edu/~hunter/m125b/ch2.pdf) that states $ \sup(A + B) = \sup(A) + \sup(B)$, we can finally write:
$$\|x + z\|^{'} = \sup\left\{\left|y^{\star}x \right|:\left\|y\right\|=1\right\} \
                + \sup\left\{\left|y^{\star}z \right|:\left\|y\right\|=1\right\}$$
Thus, by the definition of dual norm, we have:
$$\left\| x + z \right\|^{'} \le \left\| x \right\|^{'} + \left\| z \right\|^{'}$$

After considering all three conditions and making sure that the necessary conditions are met, we can say that this function is the norm.


**Second part**

To find the dual norm of the $ \ell_p $ norm in $ \mathbb{C}^N $, we need to determine the norm $ \|\cdot\|' $ defined by:
$$
\|x\|' = \sup \left\{ |y^{\star} x| : \|y\|_p = 1 \right\}.
$$
The goal is to express $ \|x\|' $ in terms of a known norm, specifically identifying which $ \ell_q $ norm corresponds to the dual of the $ \ell_p $ norm. 

For the $ \ell_p $ norm on $ \mathbb{C}^N $, it is known that the dual norm is the $ \ell_q $ norm, where $ q $ satisfies the following relationship:
$$
\frac{1}{p} + \frac{1}{q} = 1.
$$
This relationship, where $ p $ and $ q $ are Hölder conjugates, is essential in identifying the dual norm. Here’s how we can establish that the dual of the $ \ell_p $ norm is indeed the $ \ell_q $ norm:

1. **Define the $ \ell_p $ Norm**:
   For a vector $ x = (x_1, x_2, \dots, x_N) \in \mathbb{C}^N $, the $ \ell_p $ norm is defined as:
   $$
   \|x\|_p = \left( \sum_{i=1}^N |x_i|^p \right)^{\frac{1}{p}}.
   $$
   
2. **Characterize the Dual Norm**:
   By the definition of the dual norm, we have:
   $$
   \|x\|' = \sup \left\{ |y^{\star} x| : \|y\|_p = 1 \right\}.
   $$
   The expression $ |y^{\star} x| $ is the absolute value of the inner product between $ x $ and $ y $, which can be written as:
   $$
   |y^{\star} x| = \left| \sum_{i=1}^N \overline{y_i} x_i \right|.
   $$
   
3. **Apply Hölder's Inequality**:
   To simplify this, we use **Hölder's inequality**, which states that for any vectors $ x $ and $ y $ in $ \mathbb{C}^N $ and for Hölder conjugates $ p $ and $ q $  
   (with $ \frac{1}{p} + \frac{1}{q} = 1 $):
   $$
   \left| \sum_{i=1}^N \overline{y_i} x_i \right| \leq \|x\|_q \|y\|_p.
   $$
   Thus, for $ \|y\|_p = 1 $, this inequality implies:
   $$
   \sup \left\{ |y^{\star} x| : \|y\|_p = 1 \right\} \leq \|x\|_q.
   $$

4. **Achieving the Supremum**:
   We can attain equality in Hölder's inequality when $ y $ is chosen so that:
   $$
   y_i = \frac{\overline{x_i} |x_i|^{q-2}}{\|x\|_q^{q-1}} \quad \text{for } i = 1, \dots, N.
   $$
   This choice of $ y $ satisfies $ \|y\|_p = 1 $ and maximizes $ |y^{\star} x| $ to equal $ \|x\|_q $.

Therefore, we conclude that:
$$
\|x\|' = \|x\|_q,
$$
where $ q $ is the Hölder conjugate of $ p $, satisfying $ \frac{1}{p} + \frac{1}{q} = 1 $.

The dual norm of the $ \ell_p $ norm on $ \mathbb{C}^N $ is the $ \ell_q $ norm, where $ q $ is the Hölder conjugate of $ p $. Specifically:
$$
\|x\|_p' = \|x\|_q \quad \text{where } \frac{1}{p} + \frac{1}{q} = 1.
$$

Special Cases
1. **When $ p = 1 $:** The dual norm is the $ \ell_\infty $ norm.
2. **When $ p = 2 $:** The dual norm is again the $ \ell_2 $ norm, as $ p = q = 2 $.
3. **When $ p = \infty $:** The dual norm is the $ \ell_1 $ norm.

### Problem 1.3 (5 pts)

Suppose $f(x)$ is a differentiable function receiving a vector input $x = [x_i]_{i=1}^d$ and returning a scalar. Then object $\nabla_{x} f(x) = [\frac{d}{dx_i}f(x)]_{i=1}^d$ is called gradient.

Find gradient $\nabla_{x} f(x)$ for:
- (1 pts) $f(x) = (a, x)$
- (1 pts) $f(x) = \|Ax - b\|_2^2$
- (1 pts) $f(x) = \|x\|_2^2$
- (1 pts) $f(x) = \|x\|_2$
- (1 pts) $f(x) = \log_e(1 + \exp((a, x)))$

**Solution:**

In this case I want use the [definition of gradient](https://en.wikipedia.org/wiki/Gradient): 

$$
\nabla f(x) = \left(\frac{\partial f}{\partial x_1}, \
                    \frac{\partial f}{\partial x_2}, \cdots , \
                    \frac{\partial f}{\partial x_n}  \right)
$$

In the context of a scalar field $f: \mathbb{R}^n \rightarrow \mathbb{R}$, the gradient of $f$, denoted as $\nabla f$ id a vector that contains all the partial derivatives of $f$. When we consider a small change in the input variable, represented by the vector $dx = (dx_1, dx_2, \cdots, dx_n)$, the total differential $df$ can be expressed as:

$$
df = \left\langle \nabla f, dx \right\rangle  
$$


1. **Case when $f(x) = (a, x)$**

To find the gradient $ \nabla_x f(x) $ of the function $ f(x) = (a, x) $, where $ (a, x) $ denotes the dot product between vectors $ a \in \mathbb{R}^d $ and $ x \in \mathbb{R}^d $, I want to use the differential property given:

$$
df = \left\langle \nabla f, dx \right\rangle.
$$


i. **Define the Differential $ df $:**

   The function $ f(x) = (a, x) $ is defined as the dot product of $ a $ and $ x $:
   $$
   f(x) = \sum_{i=1}^d a_i x_i.
   $$
   We want to compute the differential $ df $, which represents the infinitesimal change in $ f $ due to an infinitesimal change in $ x $.

ii. **Compute the Differential $ df $ with Respect to $ x $:**

   Using the linearity of the differential, we can write:
   $$
   df = \sum_{i=1}^d a_i \, dx_i.
   $$
   This expression represents the change in $ f $ when $ x $ changes by an infinitesimal amount $ dx $.

iii. **Use the Property $ df = \left\langle \nabla f, dx \right\rangle $:**

   By the property $ df = \left\langle \nabla f, dx \right\rangle $, we can interpret $ df $ as the dot product of the gradient $ \nabla f $ and the differential $ dx $:
   $$
   df = \left\langle \nabla f, dx \right\rangle = \sum_{i=1}^d \frac{\partial f}{\partial x_i} \, dx_i.
   $$
   Comparing this with our expression $ df = \sum_{i=1}^d a_i \, dx_i $, we can see that each partial derivative $ \frac{\partial f}{\partial x_i} $ must equal $ a_i $.

iv. **Identify the Gradient $ \nabla_x f(x) $:**

   Since each partial derivative $ \frac{\partial f}{\partial x_i} = a_i $, we can write the gradient $ \nabla_x f(x) $ as:
   $$
   \nabla_x f(x) = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \\ \vdots \\ \frac{\partial f}{\partial x_d} \end{bmatrix} = \begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_d \end{bmatrix} = a.
   $$

Thus, the gradient of $ f(x) = (a, x) $ with respect to $ x $ is:
$$
\nabla_x f(x) = a.
$$


2. **Case when $f(x) = \|Ax - b\|_2^2$**

To find the gradient $ \nabla_x f(x) $ for the given function:


i. **Rewrite $ f(x) $ in terms of an inner product**:

   The squared Euclidean norm $ \|Ax - b\|_2^2 $ can be expanded as:
   $$
   f(x) = \|Ax - b\|_2^2 = (Ax - b)^{\top} (Ax - b).
   $$

ii. **Expand the Expression**:

   Expanding the expression for $ f(x) $ using the distributive property:
   $$
   f(x) = x^{\top} A^{\top} A x - 2 b^{\top} A x + b^{\top} b.
   $$
   $ b^{\top} b $ is a constant with respect to $ x $, so it will disappear while differentiation.

iii. **Compute the Differential $ df $:**

   To find the gradient, compute the differential $ df $, which represents the infinitesimal change in $ f(x) $ with respect to $ x $:
   $$
   df = d\left( x^{\top} A^{\top} A x - 2 b^{\top} A x \right).
   $$
   Let’s differentiate each term with respect to $ x $:
   
   - For the term $ x^{\top} A^{\top} A x $ , where product of $ A^{\top} A $ is symmetric matrix. I can use the fact that the differential of $ x^{\top} M x $ (where $ M $ is a symmetric matrix) with respect to $ x $ is $ 2 M x $:
     $$
     d \left( x^{\top} A^{\top} A x \right) = 2 A^{\top} A x \, dx.
     $$
   
   - To compute the differential for the term $ -2 b^{\top} A x $, first let's analyze what is it:

        - $ b^{\top} $ is the transpose of a vector $ b \in \mathbb{R}^m $,
        - $ A \in \mathbb{R}^{m \times d} $ is a matrix,
        - $ x \in \mathbb{R}^d $ is a vector of variables with respect to which we are differentiating.

        This term $ -2 b^{\top} A x $ is a *scalar-valued* function of $ x $, as it is a dot product resulting in a scalar. Then, we can rewrite it in inner product form:

        $$
        -2 b^{\top} A x = -2 \langle b, A x \rangle
        $$

        To find the differential of this term with respect to $ x $, we want to understand how a small change $ dx $ in $ x $ affects $ -2 b^{\top} A x $.

        Let’s use the **linearity of the differential**. For a term of the form $ c^{\top} x $ (where $ c $ is a constant vector with respect to $ x $), the differential is simply $ c \, dx $. This is because $ c^{\top} x $ changes linearly in $ x $, so:
        $$
        d(c^{\top} x) = c \, dx.
        $$

        We can apply this principle here because $ -2 b^{\top} A $ acts like a constant vector with respect to $ x $:
        $$
        d(-2 b^{\top} A x) = -2 (A^{\top} b) \, dx.
        $$

        Notice that:
        $$
        A^{\top} b \in \mathbb{R}^d
        $$
        is a vector in the same space as $ x $, which arises from multiplying $ A^{\top} \in \mathbb{R}^{d \times m} $ with $ b \in \mathbb{R}^m $. Therefore, $ A^{\top} b $ acts as the "coefficient" vector in the differential.

        Thus, we conclude that:
        $$
        d(-2 b^{\top} A x) = -2 A^{\top} b \, dx.
        $$
        

   Putting these together, we get:
   $$
   df = \left( 2 A^{\top} A x - 2 A^{\top} b \right) \, dx.
   $$

iv. **Identify the Gradient $ \nabla_x f(x) $:**

   By the property $ df = \left\langle \nabla f, dx \right\rangle $, we see that the gradient $ \nabla_x f(x) $ must satisfy:
   $$
   \nabla_x f(x) = 2 A^{\top} A x - 2 A^{\top} b.
   $$

Thus, the gradient of $ f(x) = \|Ax - b\|_2^2 $ with respect to $ x $ is:
$$
\nabla_x f(x) = 2 A^{\top} (Ax - b).
$$

This is a compact form that shows the gradient in terms of the residual $ Ax - b $, which is often useful in optimization contexts such as least squares problems.

3. **Case when $f(x) = \left\| x \right\|^2_2$**

Solution Steps:

i. **Express $ f(x) = \|x\|_2^2 $ by the definition**:

   The function $ f(x) = \|x\|_2^2 $ represents the squared Euclidean norm of the vector $ x \in \mathbb{R}^d $. By definition of the Euclidean norm, we can expand $ f(x) $ as:

   $$
   f(x) = \|x\|_2^2 = x^{\top} x = \sum_{i=1}^d x_i^2.
   $$

ii. **Compute the Differential $ df $ with Respect to $ x $:**

   To find the gradient, let’s calculate the differential $ df $, which represents the infinitesimal change in $ f(x) $ with respect to $ x $. For $ f(x) = x^{\top} x $, we use the fact that the differential of $ x^{\top} x $ with respect to $ x $ is $ 2 x^{\top}, dx $. And it could be rewritten in inner product term:
   $$
   df = d(x^{\top} x) = 2 x^{\top} \, dx = \left\langle 2x, dx \right\rangle
   $$
  
iii. **Identify the Gradient $ \nabla_x f(x) $:**

   By the property $ df = \langle \nabla f, dx \rangle $, we can interpret $ df = \left\langle 2x, dx \right\rangle $ as the dot product of the gradient $ \nabla f(x) $ and the differential $ dx $. Therefore, we conclude that:
   $$
   \nabla_x f(x) = 2x.
   $$

Thus, the gradient of $ f(x) = \|x\|_2^2 $ with respect to $ x $ is:
$$
\nabla_x f(x) = 2x.
$$

4. **Case when $f(x) = \left\| x \right\|_2 $**

Solution step-by-step:

i. **Look at the function $ f(x) = \|x\|_2 $:**

   The function $ f(x) = \|x\|_2 $ is the Euclidean (or $ \ell_2 $) norm of $ x \in \mathbb{R}^d $. It can be expressed as:
   $$
   f(x) = \|x\|_2 = \sqrt{x^{\top} x} = \sqrt{\sum_{i=1}^d x_i^2}.
   $$
   
ii. **Compute the Differential $ df $ with Respect to $ x $:**

   To find the gradient, we need to compute the differential $ df $. Since $ f(x) = \sqrt{x^{\top} x} $, let’s apply the chain rule for differentiation.

   We can rewrite $ f(x) $ in a form suitable for the chain rule:
   $$
   f(x) = (x^{\top} x)^{1/2}.
   $$
   Now, applying the chain rule, we differentiate $ f(x) $ with respect to $ x $:
   $$
   df = \frac{1}{2} (x^{\top} x)^{-1/2} \cdot d(x^{\top} x).
   $$

iii. **Compute $ d(x^{\top} x) $:**

   We know from the previous steps, that:
   $$
   d(x^{\top} x) = 2 x^{\top} \, dx.
   $$
   Substituting this into our expression for $ df $, we get:
   $$
   df = \frac{1}{2} (x^{\top} x)^{-1/2} \cdot 2 x^{\top} \, dx.
   $$
   Simplifying, we find:
   $$
   df = \frac{1}{2} (x^{\top} x)^{-1/2} \cdot 2 x^{\top} \, dx = \
        \frac{x^{\top} dx}{\sqrt{\left\langle x, x \right\rangle}} = \
        \frac{x^{\top} dx}{\left\| x \right\|_2} = \
        \left\langle \frac{x}{\left\| x \right\|_2}, dx \right\rangle
   $$

iv. **Identify the Gradient $ \nabla_x f(x) $:**

   By the property $ df = \langle \nabla f, dx \rangle $, we interpret $ df = \left\langle \frac{x}{\left\| x \right\|_2}, dx \right\rangle$ as the dot product of the gradient $ \nabla f(x) $ with $ dx $. Thus, we conclude that:
   $$
   \nabla_x f(x) = \frac{x}{\|x\|_2}.
   $$

Thus, the gradient of $ f(x) = \|x\|_2 $ with respect to $ x $ is:
$$
\nabla_x f(x) = \frac{x}{\|x\|_2}.
$$

5. **Case when $f(x) = \log_e(1 + \exp((a, x)))$**

Solution Steps

i. **$ f(x) $ and Chain Rule:**

   We start with:
   $$
   f(x) = \log_e(1 + \exp((a, x))).
   $$
   To find $ df $, we need to use the chain rule because $ f(x) $ is a composition of functions: first the exponential, then the addition, and finally the logarithm.

ii. **Differentiate the Outer Logarithmic Function:**

   Let’s introduce an intermediate variable $ u = (a, x) $, so that:
   $$
   f(x) = \log_e(1 + \exp(u)),
   $$
   where $ u = (a, x) $.

   Taking the differential of $ f $ with respect to $ u $, we get:
   $$
   \frac{df}{du} = \frac{1}{1 + \exp(u)} \cdot \exp(u) = \frac{\exp(u)}{1 + \exp(u)}.
   $$

   This expression can be simplified, using the properties of degree as:
   $$
   \frac{df}{du} = \frac{1}{1 + \exp(-u)}.
   $$

   This is often recognized as the **logistic function**:
   $$
   \frac{df}{du} = \sigma(u),
   $$
   where $ \sigma(u) = \frac{1}{1 + \exp(-u)} $.

iii. **Apply the Chain Rule to Find $ df $ with Respect to $ x $:**

   Now, using $ u = (a, x) $, we have:
   $$
   df = \frac{df}{du} \, du = \sigma((a, x)) \, d(a, x).
   $$

iv. **Differentiate $ (a, x) $ with Respect to $ x $:**

   Since $ (a, x) = \sum_{i=1}^d a_i x_i $, the differential $ d(a, x) $ with respect to $ x $ is:
   $$
   d(a, x) = a^{\top} \, dx.
   $$

   Therefore, we get:
   $$
   df = \sigma((a, x)) \, a^{\top} \, dx = \
        \left\langle \sigma((a, x)) \, a, dx \right\rangle.
   $$

v. **Identify the Gradient $ \nabla_x f(x) $:**

   By the property $ df = \langle \nabla f, dx \rangle $, we see that:
   $$
   \nabla_x f(x) = \sigma((a, x)) \, a.
   $$

Thus, the gradient of $ f(x) = \log_e(1 + \exp((a, x))) $ with respect to $ x $ is:
$$
\nabla_x f(x) = \frac{a}{1 + \exp(-(a, x))}.
$$

````python
def manhattan_distance(x1, x2):
    '''
    Find the Manhattan distance between two sets of points
        x1 - array of shape (n_samples, n_features)
        x2 - array of shape (m_samples, n_features)

    Returns:
        distance - array of shape (n_samples, m_samples)
    '''
    return np.sum(np.abs(x1[:, np.newaxis, :] - x2[np.newaxis, :, :]), axis=2)

def euclidean_distance(x1, x2):
    '''
    Find the Eucledian distance between two sets of points
        x1 - array of shape (n_samples, n_features)
        x2 - array of shape (m_samples, n_features)

    Returns:
        distance - array of shape (n_samples, m_samples)
    '''
    return np.sqrt(((x1[:, np.newaxis, :] - x2[np.newaxis, :, :]) ** 2).sum(2))

def cosine_distance(x1, x2):
    '''
    Find the Cosine distance between two sets of points
        x1 - array of shape (n_samples, n_features)
        x2 - array of shape (m_samples, n_features)

    Returns:
        distance - array of shape (n_samples, m_samples)
    '''

    return 1 - (x1 @ x2.T / (np.sqrt((x1 ** 2).sum(axis=1, keepdims=True)) *
                            (np.sqrt((x2 ** 2).sum(axis=1, keepdims=True)).T)))
````

### Problem 1. Frobenius Norm computation (10 pts)

1. Let $U \in \mathbb{R}^{n \times r}$ be an $n \times r$ matrix with orthonormal columns $U^{\top} U = I_r$. Let $x$ be an $m \times r$ matrix.

+ (3 pts). Propose an algorithm for the computation of $\Vert A \Vert^2_F$, where $A = U V^{\top}$, and estimate its complexity with respect to $n, m, r$. How orthogonalization can be used?

+ (3 pts). Consider the matrix $B = A \circ A$, where $\circ$ is the elementwise product of matrices (i.e., the elements of the matrix $B$ are squares of the elements of the matrix $A$). What is the maximal possible rank of the matrix $B$?

+ (5 pts). Propose an algorithm for the computation of $\Vert B \Vert^2_F$ and estimate its complexity with respect to $n, m, r$.

Solution:

**1.** To compute $ \| A \|_F^2 $ where $ A = UV^{\top} $, let's go iteratively:

i. **Condition understanding**

Given:
- $ U \in \mathbb{R}^{n \times r} $ with orthonormal columns ($ U^{\top} U = I_r $), meaning $ U $ acts as a basis with $ r $ orthonormal vectors.
- $ x \in \mathbb{R}^{m \times r} $, an arbitrary matrix.

We want to calculate $ \| A \|_F^2 $, where $ A = U x^{\top} $, using properties of orthonormal matrices and the Frobenius norm.

ii. **Simplify $ \| A \|_F^2 $ Using Properties of Frobenius Norm**

The Frobenius norm $ \| A \|_F $ of a matrix $ A $ is defined as:
$$
\| A \|_F = \sqrt{\sum_{i,j} |A_{ij}|^2}
$$
or equivalently, 
$$
 \| A \|_F^2 = \operatorname{trace}(A^{\top} A) 
$$

Thus,
$$
\| A \|_F^2 = \operatorname{trace}((UV^{\top})^{\top} (UV^{\top})) = \operatorname{trace}(x U^{\top} U x^{\top}).
$$

Since $ U $ has orthonormal columns, $ U^{\top} U = I_r $. Therefore, this fact leads us to:
$$
\| A \|_F^2 = \operatorname{trace}(x I_r x^{\top}) = \operatorname{trace}(x x^{\top}).
$$

iii. **Calculation of $ \operatorname{trace}(x x^{\top}) $**

Notice that $ x x^{\top} $ is an $ m \times m $ matrix, but we only need its trace, which can be calculated as:
$$
\operatorname{trace}(x x^{\top}) = \| x \|_F^2,
$$

This means we can compute $ \| A \|_F^2 $ by the squared norms of $ x $, avoiding the direct computation of $ A = UV^{\top} $ entirely.

iv. **Algorithm Formulation**

- **Input**: Matrices $ U \in \mathbb{R}^{n \times r} $ and $ x \in \mathbb{R}^{m \times r} $ with $ U^{\top} U = I_r $.
- **Compute**: Instead of computing $ \| A \|_F^2 $, compute $ \| x \|_F^2 $
   $$
   \| A \|_F^2 = \| x \|_F^2, \ \text{if exist} \ A = UV^{\top}
   $$

This avoids forming the matrix $ A = UV^{\top} $, which would require more operations if done explicitly.

v. **Analyze of Сomplexity**

- When calculating complexity of $ \| A \|^2_F $ requires $ O(mn) $ operations (because $A$ it's a product of two matrices of size $n \times r$ and $r \times m$), calculating complexity of $\| x \|^2_F$ requires $O(mr)$ operations.

Thus, the complexity of this approach is $ O(mr) $, which is efficient compared to directly calculations $ A $.

**2.** To find maximum possible rank of $B$ 

**i. Find decomposition of $B$**

Let's denote $r$, such that $ \text{rank}(A) = r $. From the task conditions we know, that $A = U V^{\top}$. So, we can consider decomposition of $A$ in a sum of $r$ rank-1 matrices:

$$
A = \sum_{i=1}^r u_i v_i^\top \ \ \text{or} \ \ a_{ij} = \sum_{k=1}^{r} u_{ik} v_{jk}
$$

then we can write:
$$
b_{ij} = a_{ij}^2 = \left( \sum_{k=1}^{r} u_{ik} v_{jk} \right)^2
$$

**ii. Expanding Hadamard product**

Then Hadamard product could be written as:

$$
B = A \circ A = \sum_{i=1}^r u_i v_i^\top \circ \sum_{j=1}^r u_j v_j^\top
$$

Or, expanding this explicitly:

$$
B = A \circ A = \sum_{i=1}^r \sum_{j=1}^r (u_i v_i^\top \circ u_j v_j^\top)
$$

and we obtain that:

$$
B = A \circ A = \sum_{i=1}^{r} \sum_{j=1}^r (u_i \circ u_j)(v_i \circ v_j)^\top
$$

**iii. Tricky moment with pairs**

Without any restriction, we can consider, that there are $r \times r = r^2$ possible pairs of $(k, l)$. But:
$$
\text{The term} \ \left(u_i \circ u_j\right)\left(v_i \circ v_j\right)^{\top} 
\text{is the same as} \ \left(u_j \circ u_i\right)\left(v_j \circ v_i\right)^{\top}
$$

Because element-wise multiplication is commutative. Therefore this pairs provides the same result. To avoid counting the same term twice, we consider only pairs where $i \leq j$. 

**iv. Counting pairs**

So, how to count number of unique pairs? Let's write mathematic induction for our case:
$$
\text{for} \ k=1, l \ \text{ranges from} \ 1 \ \text{to} \ r \to r \ \text{pairs} \\
\text{for} \ k=2, l \ \text{ranges from} \ 2 \ \text{to} \ r \to r - 1 \ \text{pairs} \\
\cdots \\
\text{for} \ k=r, l = r \to 1 \ \text{pair}
$$

Here we can discover arithmetic progression:
- where first element is $r$, denoting $r$-unique pairs, $a_1 = r$;
- last element is only one unique pair, $a_n = 1$;
- and the number of pairs is $r$, $n = r$

Then, sum of finite arithmetic progression, which means the number of unique pairs and therefore, maximum possible $rank(B)$ for our case, can be calculated as:
$$
S = \frac{n(a_1 + a_n)}{2} = \frac{r(r + 1)}{2}
$$ 

Thus, for $B = A \circ A$:
$$
rank(B) \leq \frac{r(r + 1)}{2}
$$

**3.** Algorithm to compute Frobenius


The Frobenius norm of a matrix $ B \in \mathbb{R}^{n \times m} $ is defined as:

$$
\|B\|_F^2 = \sum_{i=1}^n \sum_{j=1}^m (b_{ij})^2
$$

According to previous steps we can rewrite this as:

- $ b_{ij} = (a_{ij})^2 $
- $ a_{ij} = \sum_{k=1}^r u_{ik} v_{jk}$ - this operation will take $O(r)$
- $ b_{ij} = (a_{ij})^2 = \left( \sum_{k=1}^r u_{ik} v_{jk} \right)^2$ - because we already have computed this equation, and we only need to square it O(1), we still have complexity $O(r)$
- $(b_{ij})^2 = (a_{ij})^4 = \left( \sum_{k=1}^r u_{ik} v_{jk} \right)^4$ - again, the component we already have is squaring O(1), so we still have $O(r)$

For the square of Frobenius norm of $ B $:

$$
\|B\|_F^2 = \sum_{i=1}^n \sum_{j=1}^m (b_{ij})^2
$$

Adding each b_{ij} to the sum is an $O(1)$ operation for each $b_{ij}$. But, there are $n \cdot m$ elements. The total complexity for computing this step is $O(n \cdot m)$ 


Therefore, overall computational complexity is $O(nmr)$.


Below is algorithm to compute Frobenius norm  
Input: Matrix dimensions n, m, rank r, matrices U (size n x r), V (size m x r)  
Output: Frobenius norm squared of matrix $B: \left\| B \right\|_F^2$

```text
1. Initialize norm_squared ← 0
2. For i ← 1 to n do:
3.     For j ← 1 to m do:
4.         Initialize a_ij ← 0
5.         For k ← 1 to r do:
6.             a_ij ← a_ij + U[i][k] * V[j][k]
7.         End For
8.         a_ij ← a_ij - O(r)  # Adjust with computational overhead term
9.         b_ij ← a_ij^2       # Compute b_ij as the square of a_ij
10.        norm_squared ← norm_squared + b_ij
11.    End For
12. End For
13. Return norm_squared
```


### Problem 2. Unitary and orthogonal matrices (15 pts)

+ (3 pts). Let $O\in\mathbb{R}^{n\times n}$ be orthogonal matrix. Characterise explicitly all orthogonal matrices that are positive definite. You may start with $n=2$.

+ (3 pts). Show that any unitary matrix from $\mathbb{C}^{n\times n}$ can be represented as a product of at most $n$ Householder reflectors.

+ (5 pts). Give an example of $M\in\mathbb{R}^{n\times n}$ matrix that preserves $l_\infty$ norm but does not preserve $l_2$ norm; or prove that it is not possible. *Hint: you can try working with basis vectors*

+ (6 pts). Do the same for $l_1$ norm.

Solution:


**i. Recall the definition**

An orthogonal matrix $ O \in \mathbb{R}^{n \times n} $ satisfies the following:
1. $ O^T O = I $, where $ O^T $ is the transpose of $ O $ and $ I $ is the identity matrix.
2. The columns (and rows) of $ O $ are orthonormal vectors.
3. The determinant of $ O $ is either $ +1 $ or $ -1 $.

A matrix $ O $ is positive definite if for any nonzero vector $ x \in \mathbb{R}^n $, $ x^T{\top} O x > 0 $.

**ii. The limitation over eigenvalues**

By the definition of eigenvalues we can write:
$$
Ox = \lambda x
$$

Then we can write:
$$
\begin{aligned}
& Ox = \lambda x \ \ \ \ (\text{Transpose both sides}) \\
& x^{\top}O^{\top} = \lambda x^{\top} \ \ \ \ (\text{Multiply both sides by x})\\
& x^{\top}O^{\top}x = \lambda x^{\top}x \to \lambda = \frac{x^{\top}O^{\top}x}{x^{\top}x}
\end{aligned}
$$

Here, $x^{\top}x = \left\| x \right\|^2 \gt 0$, since $x \neq 0$ and $x^{\top}O^{\top}x > 0$ when $O$ is positive definite. Therefore, for such matrices $\lambda \gt 0$. Moreover, $\lambda \in \mathbb{R}^n$ in our case. 

**ii. Case $ \mathbb{R}^{2 \times 2} $**

The general form of a $ 2 \times 2 $ orthogonal matrix $ O $ can be expressed as:
$$
O = \begin{pmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{pmatrix},
$$
where $ \theta $ is the rotation angle. This represents a rotation matrix, and such matrices are orthogonal with determinant $ +1 $.

To find eigenvalues:

- Set up the characteristic equation:

   - The eigenvalues $ \lambda $ satisfy:
     $$
     \det(R - \lambda I) = 0
     $$
     where $ I $ is the identity matrix.

- Compute $ R - \lambda I $:

   $$
   R - \lambda I = \begin{bmatrix}
   \cos \theta - \lambda & -\sin \theta \\
   \sin \theta & \cos \theta - \lambda \\
   \end{bmatrix}
   $$

- Compute the determinant:

   $$
   \det(R - \lambda I) = (\cos \theta - \lambda)^2 + \sin^2 \theta
   $$

Give an example of $M\in\mathbb{R}^{n\times n}$ matrix that preserves $l_\infty$ norm but does not preserve $l_2$ norm; or prove that it is not possible. *Hint: you can try working with basis vectors*

**i. Recall the definition**

The $ l_\infty $ norm of a vector $ x \in \mathbb{R}^n $ is defined by:

$$
\|x\|_\infty = \max_i |x_i|
$$

A linear operator $ M $ preserves the $ l_\infty $ norm if it satisfies the following condition for all vectors $ x $ in $ \mathbb{R}^n $:

$$
\|Mx\|_\infty = \|x\|_\infty
$$

**ii. Supposition**

To maintain the $ l_\infty $ norm, the operator $ M $ must either rearrange the components of $ x $ or change their signs. This behavior is characteristic of matrices that have exactly one non-zero entry, either $ +1 $ or $ -1 $, in each row and each column. Such matrices are known as sign-change permutation matrices.

These sign-change permutation matrices are orthogonal, meaning that $ P^T P = I $, where $ P^T $ is the transpose of $ P $ and $ I $ is the identity matrix. Additionally, permutation matrices inherently preserve the $ l_2 $ norm of vectors.

Interestingly, any linear transformation that maintains the $ l_\infty $ norm also preserves the $ l_2 $ norm. Consequently, there does not exist a matrix $ M $ that can preserve the $ l_\infty $ norm without also preserving the $ l_2 $ norm.

Same for $l_1$ norm

**i. Recall the definition**

The $ l_1 $ norm of a vector $ x \in \mathbb{R}^n $ is defined as:

$$
\|x\|_1 = \sum_{i=1}^n |x_i|
$$

A linear operator $ M $ preserves the $ l_1 $ norm if it satisfies the following condition for all vectors $ x $ in $ \mathbb{R}^n $:

$$
\|Mx\|_1 = \|x\|_1
$$

Consider the matrix $ M $ given by:

$$
M = 
\begin{pmatrix}
\frac{1}{n} & \frac{1}{n} & \cdots & \frac{1}{n} \\
\frac{1}{n} & \frac{1}{n} & \cdots & \frac{1}{n} \\
\vdots & \vdots & \ddots & \vdots \\
\frac{1}{n} & \frac{1}{n} & \cdots & \frac{1}{n}
\end{pmatrix}
$$

For any vector $ x \in \mathbb{R}^n $, the product $ Mx $ is calculated as follows:

$$
y_i = \sum_{j=1}^n M_{ij}x_j = \sum_{j=1}^n \frac{1}{n}x_j = \frac{1}{n}\sum_{j=1}^n x_j
$$

Thus, the resulting vector $ y = Mx $ can be expressed as:

$$
y = Mx = \frac{1}{n}\left(\sum_{j=1}^n x_j\right)e_1, \quad \text{where } e_1 = 
\begin{pmatrix}
1 \\
1 \\
\vdots \\
1
\end{pmatrix} \in \mathbb{R}^n
$$

**ii. Verification for $l_1$**

Calculating the $ l_1 $ norm of $ Mx $:

$$
\|Mx\|_1 = \sum_{i=1}^n \left|\frac{1}{n} \sum_{j=1}^n x_j\right|
$$

Since each component $ y_i = \frac{1}{n} \sum_{j=1}^n x_j $ is identical for all $ i $, we have $ n $ identical terms in the sum. Therefore:

$$
\|Mx\|_1 = \sum_{i=1}^n \left|\frac{1}{n} \sum_{j=1}^n x_j\right| = n \cdot \frac{1}{n} \left|\sum_{j=1}^n x_j\right| = \left|\sum_{j=1}^n x_j\right| = \|x\|_1
$$

This shows that the matrix $ M $ preserves the $ l_1 $ norm.

**iii. Verification for $l_2$**

Next, consider the $ l_2 $ norm of $ Mx $:

$$
\|Mx\|_2 = \sqrt{\sum_{i=1}^n \left(\frac{1}{n} \sum_{j=1}^n x_j\right)^2} = \sqrt{n\left(\frac{1}{n} \sum_{j=1}^n x_j\right)^2} = \frac{1}{\sqrt{n}} \left|\sum_{j=1}^n x_j\right|
$$

In contrast, the $ l_2 $ norm of $ x $ is:

$$
\|x\|_2 = \sqrt{\sum_{i=1}^n x_i^2}
$$

Comparing the two norms, we observe that:

$$
\|Mx\|_2 = \frac{1}{\sqrt{n}} \left|\sum_{j=1}^n x_j\right| \neq \|x\|_2
$$

unless $ x $ satisfies specific conditions (e.g., all components are equal). Therefore, the matrix $ M $ preserves the $ l_1 $ norm but does not preserve the $ l_2 $ norm.

### Problem 3. Orthogonal matrices (7 pts)

A square $n \times n$ matrix $A$ is said to be in **upper Hessenberg form** if $a_{i,j}=0$ for all $i,j$ with $i > j+1$. Reducing matrix to this form will be useful for optimization of QR algorithm that can be used to get eigenvalues of the matrix.

Now, we can use orthogonal matrices to reduce matrices to upper Hessenberg form. Implement your function that could do that for any $M\in\mathbb{R}^{n\times n}$ using either Householder reflections or Givens rotations.

+ (4 pts) Implement function Hessenberg_Transform which takes a real square matrix and returns transformed matrix in Hessenberg form
+ (3 pts) Varing parameter $n$ generate random matrix of size $n \times n$ and measure the computation time of Hessenberg_Transform function. Plot observed data: computation time versus $n$. Explain obtained results.

````python
def Hessenberg_Transform(A, tolerance=1e-5):
    """
    Transforms a real square matrix A into Hessenberg form using Householder transformations.
    
    Parameters:
        A (numpy.ndarray): Real square matrix of shape (n, n)
        
    Returns:
        H (numpy.ndarray): Hessenberg form of matrix A
    """
    n = A.shape[0]
    H = np.array(A, copy=True, dtype=float)

    for j in range(n - 2):
        x = H[j+1:, j]
        
        # Compute the norm of x
        norm_x = np.linalg.norm(x)
        if norm_x < tolerance:
            continue
        
        # Construct the Householder vector
        e1 = np.zeros_like(x)
        e1[0] = 1.0
        sign = -np.sign(x[0]) if x[0] != 0 else -1.0
        v = x + sign * norm_x * e1
        v /= np.linalg.norm(v)
        
        # Apply the Householder transformation to the matrix H
        H[j+1:, j:] -= 2.0 * np.outer(v, v @ H[j+1:, j:])
        H[:, j+1:] -= 2.0 * np.outer(H[:, j+1:] @ v, v)
    
    H[np.abs(H) < tolerance] = 0.0
    return H
````

````python
# check that my realization work

from scipy.linalg import hessenberg

arr_size = 4

A = np.random.randint(1, 10, (arr_size, arr_size))
res = Hessenberg_Transform(A)
print(res, '\n')
res = hessenberg(A)
print(res)
````

````python
import time
import matplotlib.pyplot as plt

sizes = np.arange(2, 1000, 20)
times = []

for n in sizes:
    A = np.random.rand(n, n)  
    start_time = time.time()
    Hessenberg_Transform(A) 
    end_time = time.time()
    times.append(end_time - start_time)  

plt.figure(figsize=(10, 6))
plt.plot(sizes, times, linestyle='-', label="Hessenberg Transform Time")
plt.title("Computation Time vs Matrix Size (n)")
plt.xlabel("Matrix Size (n)")
plt.ylabel("Computation Time (seconds)")
plt.grid()
plt.legend()
plt.show()
````

As the size of the matrix $ n $ grows, the computational effort required to transform a matrix $ A \in \mathbb{R}^{n \times n} $ into an upper Hessenberg form increases roughly proportional to $ O(n^3) $. This arises because each application of the extended Householder transformation $ H_{\text{ext}} $ involves $ O(n^2) $ operations, and the process necessitates approximately $ n $ such transformations.

The plot reveals a rapid escalation in runtime as $ n $ increases, which is consistent with the $ O(n^3) $ complexity underlying the algorithm.

## Problem 2. Neumann series (10 pts)

Neuman series of matrix $A$ is $\sum_{k=0}^{\infty} A^{k}$.

**a.** (3 pts) Suppose Neumann series for matrix $A$ converges. Demonstrate that $\sum_{k=0}^{\infty} A^{k} = (I - A)^{-1}$.

**Solution:**


**i. Consider Partial Sums**: Let $ S_n $ be the partial sum up to $ n $:
$$
S_n = \sum_{k=0}^{n} A^k

**ii. Compute $ (I - A)S_n $**:
$$
(I - A)S_n = S_n - A S_n
$$
Substitute $ S_n $ and $ A S_n $:
$$
S_n = \sum_{k=0}^{n} A^k \quad \text{and} \quad A S_n = \sum_{k=0}^{n} A^{k+1} = \sum_{k=1}^{n+1} A^k
$$
Thus,
$$
S_n - A S_n = \left( \sum_{k=0}^{n} A^k \right) - \left( \sum_{k=1}^{n+1} A^k \right) = \

             A^0 - A^{n+1} = I - A^{n+1}

Since the series converges, $ A^{n+1} \to 0 $ as $ n \to \infty $. Therefore,
$$
\lim_{n \to \infty} (I - A)S_n = \lim_{n \to \infty} (I - A^{n+1}) = I
$$
Based on this we consider:
$$
(I - A)S = I
$$
where $ S = \lim_{n \to \infty} S_n $.

**iv. Final result**:

Since $ (I - A)S = I $, we have:
$$
S = (I - A)^{-1}
$$

Thus:
$$
\sum_{k=0}^{\infty} A^{k} = (I - A)^{-1}

**Solution:**



**Gelfand's Formula**

$$

$$

where $ \rho(A) $ is the spectral radius of $ A $, and $ \|\cdot\| $ is any consistent matrix norm.

This implies that for any $ \varepsilon > 0 $, there exists an integer $ N $ such that for all $ k \geq N $:

Let's choose $ \varepsilon $ such that $ \rho(A) + \varepsilon = r $. Then, for all $ k \geq N $:

Let $B = I - A \in \mathbb{R}^{N\times N}$, $\rho(A)<1$. Consider state space with $N$ ordinary states and one terminal state $T$. Let $p_{ij}$ and $v_{ij}$ are such that $A_{ij} = p_{ij} v_{ij}$, $\pi_i = 1 - \sum_{j} p_{ij} > 0$.

Define random variable

G_{ij}(i\rightarrow \star \rightarrow \dots \rightarrow \star \rightarrow k \rightarrow T) =

**a.** (4 pts) Show that $\mathbb{E}[G_{ij}] = \left(B^{-1}\right)_{ij}$. In other words, the average weight of all walks starting from state $i$ and reaching state $j$ right before termination equals $(B^{-1})_{ij} \pi_{j}$. We know that the Neumann series converges absolutely when $\rho(A)<1$. Is this property important for your proof?

**Hint:** Use relations between: (i) $B$ and Neumann series of $A$; (ii) $A_{ij}$ and $v_{ij}$, $p_{ij}$; (iii) $A^{k}$ and random walks.

**Solution:**


To show that $\mathbb{E}[G_{ij}] = (B^{-1})_{ij}$, we'll establish a relationship between the expected value of the random variable $G_{ij}$ and the entries of the inverse matrix $B^{-1}$.

**i. What is Random Walk:**

- The matrix $A$ represents weighted transitions between states, where $A_{ij} = p_{ij} v_{ij}$.
- Each $p_{ij}$ is the probability of transitioning from state $i$ to state $j$, and $v_{ij}$ is the associated weight.
- The termination probability from state $i$ is $\pi_i = 1 - \sum_{j} p_{ij}$, which ensures the process eventually terminates.

The random variable $G_{ij}$ is non-zero only when the last state before termination is $j$. It accounts for the accumulated weight along the path from $i$ to $j$, adjusted by $1/\pi_j$ to normalize for the termination probability at $j$. 

**ii. System of Equations:**

We define $m_{ij} = \mathbb{E}[G_{ij}]$. The expected value can be expressed recursively:

$$
m_{ij} = \sum_{k} p_{ik} v_{ik} m_{kj} + \delta_{ij}, 
\begin{cases}
&1,\text{ if } i = j;\\
&0,\text{otherwise}.
\end{cases}
$$

Here, $\delta_{ij}$ is delta, which is $1$ if $i = j$ and $0$ otherwise. This equation reflects that the expected weight from $i$ to $j$ is the sum over all possible next steps $k$, considering the transition probabilities and weights, plus an adjustment when $i = j$.

In matrix notation, the above equation becomes:

$$
M = A M + I_N
$$

Where:
- $M$ is the matrix with entries $m_{ij}$.
- $A$ is the matrix with entries $A_{ij}$.
- $I_N$ is the identity matrix of size $N$.

Rewriting the equation:

$$
(I - A) M = I_N
$$

This implies:

$$
M = (I - A)^{-1} = B^{-1}
$$

**iii. Conclusion:**

Thus, $\mathbb{E}[G_{ij}] = m_{ij} = (B^{-1})_{ij}$.

**c.** (4 pts) Explain how $G_{ij}$ should be modified if one want to compute component of solution $\left(B^{-1}x\right)_{i}$ in place of an element of inverse matrix.

**Solution:**


**i. Modified Definition of $ G_i $:**

Consider defining the random variable $ G_i $ for paths starting from state $ i $ and terminating at the terminal state $ T $ after passing through some state $ j $. We define $ G_i $ as:

$$
G_i(i \rightarrow \star \rightarrow \dots \rightarrow \star \rightarrow j \rightarrow T) = \frac{x_j}{\pi_j} V(i \rightarrow \star \rightarrow \dots \rightarrow \star \rightarrow j \rightarrow T)
$$

Here:
- $ x_j $ is the $ j $-th component of the vector $ x $.
- $ V(i \rightarrow \dots \rightarrow j \rightarrow T) $ is the product of the weights $ v_{ik} $ along the path from $ i $ to $ j $.

**ii. Expected Value:**

The expected value $ \mathbb{E}(G_i) $ can be computed as follows:

$$
\mathbb{E}(G_i) = \sum_{j=1}^N \sum_{\rho: i \rightarrow \dots \rightarrow j \rightarrow T} P_\rho \pi_j \frac{x_j}{\pi_j} V(\rho)
$$

Simplifying the expression:

- The termination probability $ \pi_j $ in the numerator and denominator cancels out.

   $$
   \mathbb{E}(G_i) = \sum_{j=1}^N \sum_{\rho: i \rightarrow \dots \rightarrow j \rightarrow T} P_\rho x_j V(\rho)
   $$

- The term $ \sum_{\rho: i \rightarrow \dots \rightarrow j \rightarrow T} P_\rho V(\rho) $ represents the total weight of all paths from $ i $ to $ j $ before termination, weighted by their probabilities.

**iii. Relating to $ (B^{-1} x)_i $:**

We can write:

$$
\mathbb{E}(G_i) = \sum_{j=1}^N x_j \left( \sum_{\rho: i \rightarrow \dots \rightarrow j \rightarrow T} P_\rho V(\rho) \right)
$$

But the inner sum is exactly $ (B^{-1})_{ij} \pi_j $, as from the properties of the random walk and the matrix $ B^{-1} $:

$$
\sum_{\rho: i \rightarrow \dots \rightarrow j \rightarrow T} P_\rho V(\rho) = (B^{-1})_{ij} \pi_j
$$

However, since $ \pi_j $ was canceled earlier, we can write:

$$
\mathbb{E}(G_i) = \sum_{j=1}^N x_j (B^{-1})_{ij}
$$

Therefore:

$$
\mathbb{E}(G_i) = (B^{-1} x)_i
$$

By defining $ G_i $ as above, we've effectively transformed the problem of computing the matrix-vector product $ (B^{-1} x)_i $ into computing the expected value of the random variable $ G_i $ in our random walk framework.

This approach aligns with your method and shows how the components of the solution $ (B^{-1} x)_i $ can be obtained by modifying $ G_{ij} $ to include the vector $ x $.

````python
import numpy as np
import matplotlib.pyplot as plt

from scipy.sparse import coo_matrix
````

````python
def MC_solver(A, i, j, N_walks, eps):
    m = A.shape[0]
    inv_ij = 0

    for _ in range(N_walks):
        current_node = j
        contribution = 1.0

        while True:
            if current_node == i:
                inv_ij += contribution

            # Termination condition with probability epsilon
            if np.random.rand() < eps:
                break

            next_node = np.random.choice(m)
            v = A[current_node, next_node] * m / (1 - eps)
            contribution *= v
            current_node = next_node

    inv_ij /= N_walks

    return inv_ij
````

**b.** (6 + 6 pts) Explain theoretically, why (and when) one would expect error to drop as $1 \big / \sqrt{N}$ where $N$ is a number of samples. Confirm this for small matrices experimentally. A good way to show this explicitly is to produce a plot alike one [available in wikipedia](https://en.wikipedia.org/wiki/Monte_Carlo_integration#Example) for Monte Carlo integration.

**Solution:**


**i. Central Limit Theorem**

The Monte Carlo method is a statistical technique that relies on random sampling to compute numerical results. When estimating an expected value $ \mu $ of a random variable $ X $ with finite variance $ \sigma^2 $, we draw $ N $ independent samples $ X_1, X_2, \ldots, X_N $ and compute the sample mean:

$$
\hat{\mu}_N = \frac{1}{N} \sum_{k=1}^{N} X_k
$$

According to the Central Limit Theorem, the distribution of the sample mean $ \hat{\mu}_N $ approaches a normal distribution with mean $ \mu $ and variance $ \sigma^2 / N $ as $ N $ becomes large:

$$
\hat{\mu}_N \xrightarrow{d} \mathcal{N}\left( \mu, \frac{\sigma^2}{N} \right)
$$

This implies that the standard deviation of the estimator $ \hat{\mu}_N $ is:

$$
\text{SE}(\hat{\mu}_N) = \frac{\sigma}{\sqrt{N}}
$$

**ii. Error Behavior**

The absolute error $ E(N) $ in the Monte Carlo estimator is typically measured as the difference between the estimated value $ \hat{\mu}_N $ and the true value $ \mu $:

$$
E(N) = |\hat{\mu}_N - \mu|
$$

Due to the properties of the normal distribution, we expect that:

$$
E(N) \approx \frac{c}{\sqrt{N}}
$$

where $ c $ is a constant related to the variance $ \sigma^2 $ of the random variable $ X $. This relationship indicates that the error decreases proportionally to $ 1/\sqrt{N} $ as $ N $ increases.

**iii. Conditions for Error Reduction**

For the $ 1/\sqrt{N} $ error, certain conditions must be met:

- Finite Variance: The random variable $ X $ being sampled must have a finite variance $ \sigma^2 $. If the variance is infinite, the CLT does not apply, and the error may not decrease as $ 1/\sqrt{N} $.

- Independence: The samples $ X_k $ must be independent and identically distributed (i.i.d.).

- Proper Sampling: The sampling method must correctly represent the distribution of $ X $.

````python
np.random.seed(45)

m = 7
alpha = 0.1

G = np.random.randn(m, m)
Q = np.linalg.qr(G)[0]
A = alpha * Q
B = np.linalg.inv(np.eye(m) - A)

eps = 0.1
i = 5
j = 4

exact_value = B[i, j]

N_values = [100*n for n in range(400, 900, 10)]
errors = []

for N_walks in N_values:
    estimated_value = MC_solver(A, i, j, N_walks, eps)
    error = abs(estimated_value - exact_value)
    errors.append(error)
    print(f"N_walks: {N_walks}, Estimated: {estimated_value}, Exact: {exact_value}, Error: {error}")

plt.figure(figsize=(8,6))
plt.loglog(N_values, errors, marker='o', label='Monte Carlo Error')
plt.loglog(N_values, [1 / N**0.5 for N in N_values], linestyle='--', label=r'Reference line ($\propto 1/\sqrt{N}$)')
plt.xlabel('Number of Walks (N)')
plt.ylabel('Absolute Error')
plt.title('Error in Monte Carlo Estimate vs Number of Walks')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.show()
````

In the proposed matrix, some elements ($ C_{ij} $) can get really large, causing even more error to build up during the calculations. This is especially a problem when there’s no limit on steps, so the error just keeps growing.

We can try to solve this problem with limitation the steps of our algorithm to avoid divergence. 

````python
def MC_solver_with_lim(A, i, j, N_walks, eps, step_limit=5):
    m = A.shape[0]
    inv_ij = 0

    for _ in range(N_walks):
        current_node = j
        contribution = 1.0
        counter = 0
        
        while True:
            counter += 1
            if counter > step_limit:
                break

            if current_node == i:
                inv_ij += contribution

            # Termination condition with probability epsilon
            if np.random.rand() < eps:
                break

            next_node = np.random.choice(m)
            v = A[current_node, next_node] * m / (1 - eps)
            contribution *= v
            current_node = next_node

    inv_ij /= N_walks

    return inv_ij
````

````python
# test case 3

np.random.seed(41)

m = 200
alpha = 0.2

G = np.random.randn(m, m)
Q = np.linalg.qr(G)[0]
A = alpha * Q
B = np.linalg.inv(np.eye(m) - A)

eps = 0.1
i = 0
j = 4
N_walks = 50000

print("Test 3", B[i, j], MC_solver_with_lim(A, i, j, N_walks, eps))
````

Yes, it provides rough estimation, but it's better then initial test case. So, we can consider that solution works.
